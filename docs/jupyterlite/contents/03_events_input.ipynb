# ipygame: Events + Input
Move the mouse over the canvas or press keys (click the canvas first).

```python
%pip install ipygame
import asyncio
import ipygame as pygame
```

```python
screen = pygame.display.set_mode((420, 260))
clock = pygame.time.Clock()
pos = [210, 130]

async def loop(frames=240):
    for _ in range(frames):
        for ev in pygame.event.get():
            if ev.type == pygame.QUIT:
                return
            if ev.type == pygame.MOUSEMOTION:
                pos[0], pos[1] = ev.pos
            if ev.type == pygame.KEYDOWN and ev.key == pygame.K_ESCAPE:
                return
        screen.fill((10, 10, 12))
        pygame.draw.circle(screen, "lime", pos, 8)
        pygame.display.flip()
        await clock.tick_async(30)

await loop()
```
