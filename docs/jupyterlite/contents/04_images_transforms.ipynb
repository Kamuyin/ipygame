# ipygame: Images + Transforms
Create a surface, save/load via Pillow, then transform it.

```python
import io
import numpy as np
import ipygame as pygame
from ipygame import image, transform
```

```python
w, h = 160, 120
arr = np.zeros((h, w, 4), dtype=np.uint8)
arr[..., 0] = np.linspace(0, 255, w, dtype=np.uint8)
arr[..., 1] = np.linspace(0, 255, h, dtype=np.uint8)[:, None]
arr[..., 2] = 120
arr[..., 3] = 255
surf = pygame.surfarray.make_surface(arr)

buf = io.BytesIO()
image.save(surf, buf, "test.png")
buf.seek(0)
loaded = image.load(buf)

rot = transform.rotate(loaded, 25)
scaled = transform.smoothscale(loaded, (240, 180))
flipped = transform.flip(loaded, True, False)

screen = pygame.display.set_mode((520, 300))
screen.fill("black")
screen.blit(loaded, (20, 20))
screen.blit(rot, (220, 10))
screen.blit(scaled, (20, 140))
screen.blit(flipped, (300, 140))
pygame.display.flip()
```
