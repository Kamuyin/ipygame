# ipygame: Drawing Primitives
Showcase draw + gfxdraw and clipping.

```python
import ipygame as pygame
from ipygame import gfxdraw
```

```python
screen = pygame.display.set_mode((520, 320))
screen.fill((20, 20, 24))

screen.set_clip((20, 20, 480, 280))
pygame.draw.rect(screen, "deepskyblue", (40, 40, 120, 80), width=4)
pygame.draw.circle(screen, "orange", (220, 90), 40, width=6)
pygame.draw.ellipse(screen, "springgreen", (300, 40, 140, 90), width=3)
pygame.draw.polygon(screen, "plum", [(60, 210), (150, 250), (90, 290)])

gfxdraw.aacircle(screen, 340, 220, 45, (255, 255, 255, 200))
gfxdraw.filled_ellipse(screen, 440, 230, 40, 20, (255, 120, 120, 200))
screen.set_clip(None)

pygame.display.flip()
```
